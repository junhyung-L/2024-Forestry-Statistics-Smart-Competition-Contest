```python
df1 = df1.dropna(subset=['밤 (kg)', '복분자딸기 (kg)', '오갈피 (kg)','마 (kg)','도라지 (kg)','더덕 (kg)','생표고 (kg)'], how='all')
df1
```

```python
df_copy=df1.copy()
df_copy
```

```python
df_copy[['밤 (kg)', '복분자딸기 (kg)', '오갈피 (kg)', '마 (kg)', 
         '도라지 (kg)', '더덕 (kg)', '생표고 (kg)']] = \
    df_copy[['밤 (kg)', '복분자딸기 (kg)', '오갈피 (kg)', '마 (kg)', 
             '도라지 (kg)', '더덕 (kg)', '생표고 (kg)']].fillna(0)
```

```python
df_copy
```

```python
regions = ['강원도', '경기도', '경상남도', '경상북도', '광주광역시', '대구광역시', 
           '대전광역시', '부산광역시', '울산광역시', '인천광역시', '전라남도', '전라북도', 
           '제주', '충청남도', '충청북도']

columns_to_round = ['토양깊이유형', '토성코드', '토양형코드', '토양유효수분량']
columns_to_mean = ['연평균기온', '최고기온', '최저기온', '상대습도(%)', '강수량(mm)']

for region in regions:
    mask = df_copy['지역'].str.contains(region)
    
    df_copy.loc[mask, columns_to_round] = df_copy.loc[mask, columns_to_round].transform(
        lambda x: x.fillna(x.mean().round())
    )
    
    df_copy.loc[mask, columns_to_mean] = df_copy.loc[mask, columns_to_mean].transform(
        lambda x: x.fillna(x.mean())
    )
df_copy
```
